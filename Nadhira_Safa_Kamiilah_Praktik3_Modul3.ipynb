##Nomor 1

```python
query1="""WITH monthly_sales AS (
    SELECT strftime('%Y-%m', Order_Date) AS period, SUM(Retail_Price * Quantity) AS total_sales
    FROM sales
    INNER JOIN product ON sales.Item_id = product.Item_id
    GROUP BY period),
monthly_sales_previous AS (
    SELECT strftime('%Y-%m', Order_Date, '-1 month') AS period,
    SUM(Retail_Price * Quantity) AS prev_month_sales
    FROM sales
    INNER JOIN product ON sales.Item_id = product.Item_id
    GROUP BY period)
SELECT
    m1.period AS period,
    m1.total_sales AS total_sales,
    COALESCE(m2.prev_month_sales, 0) AS prev_month_sales,
    ROUND(((m1.total_sales - COALESCE(m2.prev_month_sales, 0)) / COALESCE(m2.prev_month_sales, 1)) * 100, 2) AS sales_growth
FROM monthly_sales m1
LEFT JOIN monthly_sales_previous m2 ON m1.period = m2.period
ORDER BY period;"""
pd.read_sql_query(query1,conn)
```

##Nomor 2

```python
query2= """WITH CustomerSpending AS (
    SELECT s.CustomerID, SUM(s.Quantity * p.Retail_Price) AS total_spending
    FROM Sales s
    JOIN Product p ON s.Item_id = p.Item_id
    JOIN Customer c ON s.CustomerID = c.CustomerID
    WHERE s.Shipping_State NOT IN ('Wyomin', 'Idaho', 'Alaska')
    GROUP BY s.CustomerID),
CustomerCategory AS (
    SELECT CustomerID,
        CASE
            WHEN total_spending > 500 THEN 'High_spender'
            WHEN total_spending BETWEEN 300 AND 500 THEN 'Mid_spender'
            ELSE 'Low_spender'
        END AS cust_cat
    FROM CustomerSpending)
SELECT cc.cust_cat AS customer_category,
    COUNT(*) AS total_users,
    AVG(s.Quantity) AS avg_quantity,
    AVG(cs.total_spending) AS avg_spending
FROM CustomerCategory cc
JOIN CustomerSpending cs ON cc.CustomerID = cs.CustomerID
JOIN Sales s ON cc.CustomerID = s.CustomerID
GROUP BY cc.cust_cat;"""
pd.read_sql_query(query2,conn)
```

##Nomor 3

```python
query3="""WITH product_ratings AS (
    SELECT p.Item_id, p.Description, AVG(s.Rating) AS avg_rating
    FROM sales s
    JOIN product p ON s.Item_id = p.Item_id
    WHERE strftime('%Y', s.Order_Date) = '2004'
    GROUP BY p.Item_id, p.Description),
top_products AS (
    SELECT Item_id, Description, avg_rating, RANK() OVER (ORDER BY avg_rating DESC) AS Ranking
    FROM product_ratings
    ORDER BY avg_rating DESC
    LIMIT 5),
bottom_products AS (
    SELECT Item_id, Description, avg_rating, RANK() OVER (ORDER BY avg_rating ASC) AS Ranking
    FROM product_ratings
    ORDER BY avg_rating ASC
    LIMIT 5)
SELECT '2004' AS Year, p1.Description AS Product, p1.avg_rating, p1.Ranking
FROM top_products p1
UNION ALL
SELECT
    '2004' AS Year, p2.Description AS Product, p2.avg_rating, p2.Ranking
FROM bottom_products p2;"""
pd.read_sql_query(query3,conn)
```
